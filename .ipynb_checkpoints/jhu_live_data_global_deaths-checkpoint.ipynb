## Import all cases – UK, France, Denmark and the netherlands have overseas territory that ne need to drop from our groupby count

– Fill NaN with 'test' as the column is populated by strings

– Create two new dataframes – drop `test`, and drop `is not` test

– Drop what we don't want from the smaller `is not` set

– Join data together

https://www.science-emergence.com/Articles/How-to-remove-one-or-multiple-rows-in-a-pandas-DataFrame-in-python-/


```python
global_deaths = pd.read_csv('https://raw.githubusercontent.com/CSSEGISandData/COVID-19/master/csse_covid_19_data/csse_covid_19_time_series/time_series_covid19_deaths_global.csv',
                       parse_dates=True)
global_deaths.head()
```

```python
global_deaths = global_deaths.fillna('test')
global_deaths.head()
```

## What is in the Province/State column?

We should split these rows as we only want country data – so we drop overseas territory, and then sum China, Canada and Australia

```python
global_deaths['Province/State'].unique()
```

```python
global_deaths['Country/Region'].unique()
```

## Create new dataframe with `only countres` – we're dropping all rows in `Province/State` that are not `'test'`

```python
countries = global_deaths.drop(global_deaths[global_deaths['Province/State'] != 'test'].index)
countries.head()
```

## Now let's create a dataframe containing only the `overseas territories or regions` in Province/State – that's all the `rows marked 'test'`

```python
regional = global_deaths.drop(global_deaths[global_deaths['Province/State'] == 'test'].index)
regional.head()
```

```python
regional['Country/Region'].value_counts()
```

## Now let's drop the rows containing UK, France, Denmark and the Netherlands from the `regional dataframe`

```python
lose_these = ['United Kingdom', 'France', 'Netherlands', 'Denmark']
regional.drop(regional[regional['Country/Region'].isin(lose_these)].index, inplace=True)
```

```python
regional['Country/Region'].value_counts()
```

## So now we have a dataframe of countries called `countries` and a dataframe of countries by regional called `regional` that excludes all of the UK, France, Denmark and Netherlands overseas territories

## Let's join the data

```python
data_frames = [countries, regional]
deaths_clean =  pd.concat(data_frames)
deaths_clean
```

```python
deaths_clean['Country/Region'].value_counts()
```

## And now let's group by country to sum Australia, Canada and China regions

```python
final_deaths = deaths_clean.groupby('Country/Region').sum()
final_deaths
```

## And sort A-Z and reset index

```python
final_deaths = final_deaths.reset_index()
final_deaths = final_deaths.sort_values(by='Country/Region')
final_deaths
```

## Save and expor the final cleaned up dataset – and re-import

All countries grouped by `Country / Region`

```python
final_deaths.to_csv('data/jhu/global_deaths_clean.csv')
```

## Import cleaned data

We can't parse dates tet as the column headers are not dates

Remove index colimn in excel

```python
total_deaths = pd.read_csv('data/jhu/global_deaths_clean.csv', index_col=0,
                          usecols =[i for i in final_deaths if i != ''])
total_deaths
```

## Drop the lat and long cols – we don't need them

```python
total_deaths.drop(['Lat', 'Long'], axis=1, inplace=True)
total_deaths.head()
```

## Select cumulative countries we want to look at

```python
selected_deaths = total_deaths.loc[['United States', 'United Kingdom', 'Brazil', 'China', 
                                    'France', 'Germany', 'Iran', 'Italy', 'Mexico']]
selected_deaths
```

```python
selected_deaths = selected_deaths.transpose()
selected_deaths.to_csv('data/selected_deaths.csv')
```

## Get daily case count per country

`Use diff() method`

Axis = 1 to calculate along columns

```python
daily_deaths = total_deaths.diff(axis=1, periods=1)
daily_deaths.head()
```

```python
## Get UK daily figs as a sanity check
uk_test = daily_deaths[daily_deaths.index == 'United Kingdom']
uk_test
```

## Rolling seven day average of new cases

Use `rolling()` method

`Axis = 1` to average along columns

```python
seven_days = daily_deaths.rolling(7, axis=1).mean()
seven_days.head()
```

## % difference between two dates using `pct_change()` method

select the colums to compare

```python
final_cols = seven_days.loc[:, ['4/28/20', '5/17/20']]
final_cols.head()
```

```python
final_cols['percent_change'] = ((final_cols['5/17/20'] / final_cols['4/28/20']) - 1) * 100
final_cols.head()
```

```python
final_cols = seven_days.loc[:, ['4/28/20', '5/17/20']]
final_cols.head()
```

```python
final_cols_pct = final_cols.pct_change(axis=1) * 100
final_cols_pct.head()
```

## Transpose the daily cases data and reset the index so we can visualise after changing string into datetime object

```python
daily_deaths.head()
```

```python
transp = daily_deaths.transpose()
transp.reset_index(inplace = True) 
transp.tail()
```

## Convert date string to datetime objects

```python
transp['index'] = pd.to_datetime(transp['index'])
```

```python
fig, ax = plt.subplots(figsize=(10,6))
ax.plot(transp['index'], transp['India'].rolling(7).mean(), label='rolling ave', c='r');
ax.bar(transp['index'], transp['India'], label='new daily deaths')
ax.set_xlim('2020-03-15', '2020-06-8')
plt.title('India deaths by day, since date of first death')
plt.legend()
plt.savefig('output/jhu/india_daily_deaths_18_may.pdf');
```

```python
selected_deaths = transp[['index','United Kingdom', 'France', 'Italy', 'US', 'Mexico', 'Brazil', 'India']]
selected_deaths.tail()
```

```python
selected_deaths.to_csv('data/jhu/out_global_deaths.csv')
```

## Check data in excel, delete the index col and import

```python
new_deaths = pd.read_csv('data/jhu/out_global_deaths.csv', parse_dates=True, index_col=0, dayfirst=True,
                        usecols =[i for i in selected_deaths if i != ''])
new_deaths.tail()
```

```python
fig, ax = plt.subplots(figsize=(10, 6))

for column, value in new_deaths.items():
    ax.plot(value.rolling(7).mean(), label=column)

# plt.yscale('log')
plt.ylabel('cases')
plt.xlabel('date')
plt.title('Total deaths by day – seven day average')
plt.xticks(rotation=30, ha='right')
ax.xaxis.set_major_formatter(date_format)

plt.legend()
plt.savefig('output/jhu/global_daily_deaths.pdf')
```

## Deaths as small multiples

```python
fig, ax = plt.subplots(3, 3, figsize=(15,7), sharex=True, sharey=True)

plt.suptitle('Total deaths by day – seven day average of new cases') 

for ax, column in zip(ax.flat, new_deaths):
    ax.set_title(column)
    new_deaths[column].rolling(7).mean().plot(ax=ax)
    plt.sca(ax)
    plt.xticks(rotation=30, ha='right')
    ax.xaxis.set_major_formatter(date_format)
    ax.set_xlim('2020-02-01', '2020-06-20')
    
plt.show()
# plt.savefig('XXXX.pdf')
```
